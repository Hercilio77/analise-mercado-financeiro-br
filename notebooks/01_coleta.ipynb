# 01 - Coleta de Dados Via API

Fonte: Banco Central do Brasil (SGS - Sistema Gerenciador de Séries Temporais)
Objetivo: Coletar dados históricos de SELIC, IPCA e câmbio USD/BRL e salvar em 'data/raw/'

432 - Taxa Selic
433 - IPCA
1 - câmbio USD/BRL

```python
import requests
import pandas as pd
from datetime import datetime
from io import StringIO
```

```python
def consumir_api(url, formato, data_inicial='01/01/2020', data_final=None):
    if data_final is None:
        data_final = datetime.today().strftime("%d/%m/%Y")

    params = {
        'dataInicial': data_inicial,
        'dataFinal': data_final,
        'formato': formato
    }


    response = requests.get(url, params=params)

    if response.status_code != 200:
        print("Erro na requisição: {response.status_code}")
        return None

    
    if formato == 'json':
        return pd.DataFrame(response.json())
        
    elif formato == 'csv':
        return pd.read_csv(StringIO(response.text))
        
    else:
        print("Formato inválido")
        return None

```

```python
selic = consumir_api('https://api.bcb.gov.br/dados/serie/bcdata.sgs.432/dados','json')
```

```python
ipca = consumir_api('https://api.bcb.gov.br/dados/serie/bcdata.sgs.433/dados', 'json')
```

```python
cambio = consumir_api('https://api.bcb.gov.br/dados/serie/bcdata.sgs.1/dados', 'json')
```
